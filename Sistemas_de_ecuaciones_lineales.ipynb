## Caracterización

1. Un sistema de ecuaciones es el conjunto de dos o más ecuaciones (las cuales comparten las mismas variables, explícita o implícitamente). Ejemplo: _las dos líneas de frutas, las dos ecuaciones_.

```python
display(f1)
display(f2)
print("--------")
display(ecu1)
display(ecu2)
```

2. Un sistema de ecuaciones lineales consiste en un sistema de ecuaciones en el que el máximo exponente de las dos variables que tienen las ecuaciones es 1. Ejemplo: _$x$ tiene un exponente 1, $y$ tiene un exponente 1_

3. Resolver o solucionar un sistema de ecuaciones lineales consiste en hallar los valores de las dos variables que hacen que se cumplan ambas ecuaciones al tiempo; en otras palabras, se trata de hallar cuánto valen las variables a fin de que se cumplan las igualdades expresadas en las ecuaciones. Ejemplo: _hallar cuánto vale la manzana, hallar cuánto vale el banano; hallar cuánto vale $x$, hallar cuánto vale $y$._

```python
display(sp.Eq(manzana,10))
display(sp.Eq(banano,1))
display(sp.Eq(x,10))
display(sp.Eq(y,1))
display((10,1))
```

4. La manera de saber si las soluciones halladas son correctas, es reemplazando los valores obtenidos en las ecuaciones iniciales y verificando que se llegue a una premisa verdadera. Compruébelo: reemplace el valor $10$ en $x$ y $1$ en el valor $y$; resuelva y verifique si hay verdad en las igualdades presentadas ("True").

```python
display(ecu1)
display(ecu2)
print("------")
display(sp.Eq(3*10,30))
display(sp.Eq(10+8*1,18))
```

## Métodos de solución de sistemas de ecuaciones lineales
Existen múltiples maneras de solucionar sistemas de ecuaciones lineales (*Regla de Cramer*, *Eliminación Gauss Jordan o por matrices*, etc.). En esta lección nos centraremos en 4: de forma gráfica, por igualación, por eliminación y por sustitución.

### De forma gráfica

Si graficaramos las ecuaciones anteriores:

```python
display(ecu1)
display(ecu2)
```

Tendríamos una gráfica así que tiene dos líneas. Ambas se cruzan en el punto $(10,1)$ (compruébelo). Pues bien, encontrar la solución a un sistema de ecuaciones de forma gráfica consiste en encontrar el punto exacto en que las dos rectas se cruzan. La limitación de este método está en que hay sistemas que tienen su intersección en números decimales difíciles de estimar (este método es muy inexacto e impreciso).

## Terminología

En este punto ya sabemos resolver sistemas de ecuaciones lineales. Vamos a hablar de los últimos tres términos que necesitas saber: sistema consistente determinado, sistema consistente indeterminado y sistema inconsistente.

### Sistema consistente determinado

En las *ecuaciones de las frutas* vimos que sólo cuando $x$ vale 10 y $y$ vale 1 es que el sistema cumple las igualdades. Pues bien, eso es un sistema de ecuaciones **consistente** (que tiene solución) **determinado** (que tiene una única solución). Su gráfica (la del sistema) son dos lineas rectas que se cruzan en un único punto (el punto que representa la solución).

### Sistema consistente indeterminado 

Grafica este sistema de ecuaciones lineales:

```python
display(sp.Eq(4*x+5*y,30))
display(sp.Eq(8*x+10*y,60))
```

¿Notaste que ambas ecuaciones se superponen? Pues bien, quiere decir que técnicamente son la misma ecuación (tanto como que la segunda es el doble de la primera). En este caso, al ser la misma ecuación, el sistema comparte todos los puntos entre sí (la línea de la ecuación 1 toca todos los puntos de la línea de la ecuación 2). Por esta razón se dice que tiene infinitas soluciones. El sistema es pues uno **consistente** (tiene respuesta) **indeterminado**(no hay una única y puntual solución, sino infinitas).

En los métodos no gráficos de solución se identifica un sistema de ecuaciones consistente indeterminado por cuanto las simplifaciones son absolutas (incluso de las variables) y llegamos a expresiones sin variables que son verdad (tu profesor explicará más este punto).

### Sistema inconsistente

Grafica este sistema:

```python
display(sp.Eq(4*x+5*y,30))
display(sp.Eq(4*x+5*y,20))
```

¿Ves que no se cruzan? Pues bien, por nunca cruzarse se dice que ambas rectas son *paralelas*. Cuando las dos líneas rectas de un sistema de ecuaciones lineal son paralelas, el sistema no tiene solución (no hay puntos de cruce entre las ecuaciones del sistema). De ahí que se le diga al sistema **inconsistente** (no hay solución del sistema).

En los métodos no gráficos de solución se identifica un sistema de ecuaciones inconsistente  por cuanto las simplifaciones son absolutas (incluso de las variables) y llegamos a expresiones sin variables que son falsas -no tiene sentido- (tu profesor explicará más este punto).
